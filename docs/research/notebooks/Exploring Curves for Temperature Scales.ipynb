# Sine

```maxima
set_plot_option([svg_file, "maxplot.svg"])$
```

```maxima
plot2d(sin(x)*306, [x, %pi/4, 3*%pi/4])$
```

```maxima
phase: -0.8$
plot2d(sin(x - phase)*306, [x, %pi/4+phase, 3*%pi/4+phase])$
```

```maxima
maxY: 306$
halfWidth: %pi/4$
fullWidth: %pi/2 - halfWidth$
plot2d(sin(x + halfWidth) * maxY, [x, 0, 2 * fullWidth])$
```

# Catenary

```maxima
acosh(306), numer;
plot2d(cosh(x), [x, -%pi, %pi])$
```

```maxima
cosh(2.042508 * %pi), numer;
```

```maxima
6.416729612588352 / %pi, numer;
```

```maxima
maxY: 306;
curvature: 1;
maxX: curvature * acosh(maxY / curvature), numer;
plot2d(curvature * cosh(x / curvature), [x, -maxX, maxX])$
```

```maxima
curvature: 220;
maxX: curvature * acosh(maxY / curvature), numer;
plot2d(curvature * cosh(x / curvature), [x, -maxX, maxX])$
```

```maxima
plot2d(cosh(x), [x, -0.1, 0.1])$
```
